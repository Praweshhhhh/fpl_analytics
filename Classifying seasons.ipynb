```python
season1_start = 2018 * 365 + 8 * 30 + 1 
season1_end = 2019 * 365 + 5 * 30 + 30 

season2_start = 2019 * 365 + 8 * 30 + 1 
season2_end = 2020 * 365 + 8 * 30 + 15

season3_start = 2020 * 365 + 9 * 30 + 1 
season3_end = 2021 * 365 + 5 * 30 + 30
```

```python
path2021 = "/home/pratik/Documents/MSc/Project/Data Preparation/data/2020-21/players/"
dir_list_2021 = os.listdir(path2021)
```

```python
for i in range(len(dir_list_2021)):
    data_2021 = pd.read_csv(path2021+ dir_list_2021[i] +"/gw.csv" )
    data_2021['Season'] =  ""
    for j in range(len(data_2021)):
        print(data_2021['kickoff_time'][j])
        year = int(data_2021['kickoff_time'][j][0:4])
        month = int(data_2021['kickoff_time'][j][5:7])
        day = int(data_2021['kickoff_time'][j][8:10])
        total = year * 365 + month * 30 + day

        if total > season1_start and total < season1_end:
            data_2021['Season'][j] = "1819"
        elif total > season2_start and total < season2_end:
            data_2021['Season'][j] = "1920"
        elif total > season3_start and total < season3_end:
            data_2021['Season'][j] = "2021"
        else:
            print("___NOT FOUND___")
    data_2021.to_csv(path2021+ dir_list_2021[i] +"/gw.csv", index=False)
```
