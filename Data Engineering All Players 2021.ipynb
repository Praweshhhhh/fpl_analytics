```python
for i in range(len(dir_list_2021)):
    data_2021 = pd.read_csv(path2021 + str(dir_list_2021[i]) + "/gw.csv")
    data_2021['fixture_gap'] = ""
    
    for j in range(len(data_2021)):
        if j == 0:
            data_2021['fixture_gap'][j] = 7
        else:
            year1 = int(data_2021['kickoff_time'][j][0:4])
            month1 = int(data_2021['kickoff_time'][j][5:7])
            day1 = int(data_2021['kickoff_time'][j][8:10])
            hour1 = int(data_2021['kickoff_time'][j][11:13])
            mins1 = int(data_2021['kickoff_time'][j][14:16])


            year0 = int(data_2021['kickoff_time'][j-1][0:4])
            month0 = int(data_2021['kickoff_time'][j-1][5:7])
            day0 = int(data_2021['kickoff_time'][j-1][8:10])
            hour0 = int(data_2021['kickoff_time'][j-1][11:13])
            mins0 = int(data_2021['kickoff_time'][j-1][14:16])

            time_start = (year0 * 365 + month0 * 30 + day0) * 1440 + hour0 * 60 + mins0
            time_end = (year1 * 365 + month1 * 30 + day1) * 1440 + hour1 * 60 + mins1
            time_diff = time_end - time_start
            days_gap = time_diff / 1440
            if days_gap > 25:
                data_2021['fixture_gap'][j] = 7
            elif days_gap <= 25 and days_gap > 10:
                data_2021['fixture_gap'][j] = 5
            else:
                data_2021['fixture_gap'][j] = days_gap
    data_2021.to_csv(path2021 + str(dir_list_2021[i]) + "/gw.csv", index=False)
    
```

```python
for j in range(len(dir_list_2021)):
    data_2021 = pd.read_csv(path2021 + str(dir_list_2021[j]) + "/gw.csv")
    data_2021['goal_difference'] = ""
    data_2021['was_home_binary'] = ""
    for i in range(len(data_2021)):
        if str(data_2021['was_home'][i]) == 'True':
            data_2021['was_home_binary'][i] = 1
        else:
            data_2021['was_home_binary'][i] = 0

        if data_2021['was_home_binary'][i] == 1:
            data_2021['goal_difference'][i] = data_2021['team_h_score'][i] - data_2021['team_a_score'][i] 
        else:
            data_2021['goal_difference'][i] = data_2021['team_a_score'][i] - data_2021['team_h_score'][i]
    data_2021.to_csv(path2021 + str(dir_list_2021[j]) + "/gw.csv", index=False)
```
