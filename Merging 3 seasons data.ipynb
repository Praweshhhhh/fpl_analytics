# Merging data from 18-19,19-20 and 20-21 seasons into single csv

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import namedtuple
import random
from random import sample
from collections import Counter
import numba
import os
%matplotlib inline
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
path1819 = "/home/pratik/Documents/MSc/Project/Data Preparation/data/2018-19/players/"
path1920 = "/home/pratik/Documents/MSc/Project/Data Preparation/data/2019-20/players/"
path2021 = "/home/pratik/Documents/MSc/Project/Data Preparation/data/2020-21/players/"
dir_list_1819 = os.listdir(path1819)
dir_list_1920 = os.listdir(path1920)
dir_list_2021 = os.listdir(path2021)
# print('\n'.join(dir_list_2021))
```

```python
for item in dir_list_2021:
    if item in dir_list_1819:
        if item in dir_list_1920:
            print(item,"Yes","Yes")
        else:
            print(item, "Yes","No")
    else:
        if item in dir_list_1920:
            print(item,"No","Yes")
        else:
            print(item,"No","No")
```

```python
count = 0
for item in dir_list_2021:
    if item in dir_list_1819:
        file_1819_path = path1819 + str(item) + "/gw.csv"
        data_1819 = pd.read_csv(file_1819_path)
        file_2021_path = path2021 + str(item) + "/gw.csv"
        data_2021 = pd.read_csv(file_2021_path)
        data_1819 = data_1819.append(data_2021, ignore_index = True)
        data_1819.to_csv(file_2021_path,index=False)
#         print("1920",file_1920_path)
#         print("2021",file_2021_path)
        count +=1
        folder_path = path1819 + str(item) + "/"
        if os.path.exists(file_1819_path):
            if os.path.exists(folder_path + "history.csv"):
                os.remove(folder_path + "history.csv")
            os.remove(file_1819_path)
        if os.path.exists(folder_path):
            os.rmdir(folder_path)
```
